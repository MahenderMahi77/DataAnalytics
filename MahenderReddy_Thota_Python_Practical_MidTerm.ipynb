### Q1. Pivot Table showing average tip amount spent each day for each gender

```python
#Importing Seaborn and Pandas packages
import seaborn as sns
import pandas as pd
```

```python
#Loading the "Tips" dataset available in Seaborn package to a dataframe
tips = sns.load_dataset('tips')
```

```python
#Using Pandas head() function to see if the dataframe has right data in it
tips.head()
```

```python
#Using pivot_table() for a spreadsheet-style pivot table as a DataFrame
table = pd.pivot_table(tips,values='tip',index=['day'],columns=['sex'],aggfunc='mean')
```

```python
#Print the Pivot Table
table
```

### Q2. Women who did not survive the crash and were travelling in the third class

### Q3. Tips data frame with lowest to highest Tip Percentage

```python
#Using Pandas head() function to see if the dataframe has right data in it
tips.head()
```

```python
#Creating a new column for tip% in the dataframe as a calcualted column
tips['tip_percentage'] = (tips['tip']/tips['total_bill'])*100
```

```python
#Checking the new column values
tips.head()
```

```python
#Sorting dataframe based on tip_percentage column
tips = tips.sort_values(by=['tip_percentage'])
```

```python
#Printing the dataframe
tips
```

### Q4. Average weight of cars manufactured in each country

```python
#Loading the 'mpg' dataset available in Seaborn and checking data in dataframe
cars = sns.load_dataset('mpg')
cars.head()
```

```python
#Grouping the rows based on 'origin' and aggregation using mean()
weight = cars[['origin','weight']].groupby('origin').mean()
```

```python
#Printing the result
weight
```
