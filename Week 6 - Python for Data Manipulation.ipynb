#### Q4 a) Subset manipulations

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

###### Only return alone, alive, and sex  column

```python
titanic.loc[:, ['alone', 'alive', 'sex']]
titanic.loc[:,['alone','alive','sex']]
```

###### Only return columns from start to class column

```python
titanic.iloc[:, :9]
```

###### Only return top 10 rows and 2nd, 5th, and 6th columns

```python
titanic.iloc[:10,[2,5,6]]
```

#### Q4 b) row subsetting (filtrations) manipulations on Titanic dataset

##### 1) passenger was alone

```python
titanic[(titanic.alone == False)]
```

##### Passenger was alone and did not survive the crash

```python
titanic[
    (titanic.alone == False)
    & (titanic.survived == 0)
]
```

###### Use loc to only return Column fare for female passengers who were in 1st clas of the ship (2 conditions)

```python
titanic.loc[
    (titanic['sex'] == 'female') & (titanic['class'] == 'First'), ['fare']]
```

```python
titanic[
    (titanic.sex == 'female')
    & (titanic['class'].isin(['First']))
]
```

```python
titanic[
    (titanic.sex == 'female')
    & (titanic['class'].isin(['First', 'Third']))
    & (titanic.age > 30)
    & (titanic.survived == 0)
]
```

###### Return the 'age' and 'fare' values for row #4 in the titanic dataset

```python
titanic.loc[4, ['age', 'fare']]
```

###### Return the 'age' and 'fare' values for 2 rows: #4 and #5

```python
titanic.loc[[3, 4], ['age', 'fare']]
```

###### Example of a query using the titanic dataset

```python
titanic[
    (titanic.sex == 'female')
    & (titanic['class'].isin(['First', 'Third']))
    & (titanic.age > 30)
    & (titanic.survived == 0)
]
```
